```python
import requests as rq
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv(r'https://raw.githubusercontent.com/siddheshphatarpekar/Dataset/main/India_State_health_Budget_Record.csv')
c1 = df['State_UT'].astype(str)
c2 = df['Total Health Sector Expenditure as % of State Budget']
c3 = df['Capital Expenditure (including Loans & Advances ) as % of Total Health Sector Expenditure']
plt.style.use("ggplot")
plt.figure(figsize=(12,10))
plt.title("Health BudgetState Wise")
plt.xticks(visible='false', rotation = 90)
plt.ylabel("Capital Expenditure (including Loans & Advances )",color ='red')
plt.bar(c1,c2)
plt.show()
```
