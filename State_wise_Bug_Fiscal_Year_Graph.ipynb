```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

col_nm = ['Fiscal Year','Total Health Sector Expenditure as % of State Budget']
df = pd.read_csv(r'India_State_health_Budget_Record.csv',usecols=col_nm).dropna(subset='State_UT')
plt.style.use("fivethirtyeight")
plt.figure(figsize=(12,10))
plt.title("Health BudgetState Wise")
plt.xlabel("Fiscal Year")
label = ["State_UT"]
plt.xticks(visible='false')
plt.ylabel("Capital Expenditure (including Loans & Advances )")
plt.bar(df.State_UT,df["Total Health Sector Expenditure as % of State Budget"])
plt.show()
```
