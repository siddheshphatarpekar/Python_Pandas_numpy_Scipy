```python
import pandas as pd
import matplotlib.pyplot as plt
col=['state','percent','dollars','pay','region']
df = pd.read_csv(r'https://raw.githubusercontent.com/siddheshphatarpekar/Dataset/main/Pay%20Rate%20.csv?token=GHSAT0AAAAAAB3U3EVEWBIDV5JBVFCIQNHYY4BX6KA',usecols=col)
payhour = df["pay"].div(df["dollars"].values)
df.sort_values(by = 'region')
f = plt.figure()
f.set_figwidth(12)
f.set_figheight(10)
plt.style.use('ggplot')
plt.xlabel("State")
plt.ylabel("Pay Per Hour")
plt.bar(df.region,payhour,width=0.4,color='orange')
plt.scatter(df.region,df.dollars,color='darkblue')
plt.show()
```
