```python
import numpy as np
import string
import random
import matplotlib.pyplot as plt
arr1 = np.array(np.random.rand(7))
ch = np.random.choice(list(string.ascii_lowercase),size=7)
arr2 = np.array(ch)
for i in arr1:
    arr1 = arr1+1
    fig, ax= plt.subplots(2,2,sharey=False)
    fig.suptitle('Graph Plots Using Numpy Array')
    ax[0,0].plot(arr2,arr1)
    ax[0,1].scatter(arr2,arr1)
    ax[1,0].bar(arr2,arr1)
    ax[1,1].barh(arr2,arr1)
plt.show()
```
